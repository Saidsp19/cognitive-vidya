```python
# --- EVALUATION ENGINE ---
def evaluate():
    passed = 0
    total = 1

    def to_list(head):
        out = []
        while head:
            out.append(head.val)
            head = head.next
        return out

    def from_list(vals):
        dummy = ListNode(0)
        cur = dummy
        for v in vals:
            cur.next = ListNode(v)
            cur = cur.next
        return dummy.next

    cases = [([1,2,3,4,5], [5,4,3,2,1]), ([1,2], [2,1]), ([], [])]
    for i, (inp, expected) in enumerate(cases):
        head = from_list(inp) if inp else None
        try:
            result = to_list(reverse_list(head))
            if result == expected:
                passed += 1; print(f"✅ Test {i+1} Passed")
            else:
                print(f"❌ Test {i+1} Failed. Expected {expected}, got {result}")
        except Exception as e:
            print(f"⚠️ Test {i+1} Error: {e}")

    if passed == total:
        print("\n🎉 ALL TESTS PASSED!")
    else:
        print(f"\n📊 Score: {passed}/{total} tests passed")

evaluate()
```
