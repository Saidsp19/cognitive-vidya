```python
# --- EVALUATION ENGINE ---
def evaluate():
    passed = 0
    total = 1
    strs = ["eat","tea","tan","ate","nat","bat"]
    expected = [["bat"],["nat","tan"],["ate","eat","tea"]]
    try:
        result = group_anagrams(strs)
        norm = sorted([sorted(g) for g in result], key=lambda g: g[0] if g else "")
        exp_norm = sorted([sorted(g) for g in expected], key=lambda g: g[0] if g else "")
        if norm == exp_norm:
            passed += 1
            print("✅ Test 1 Passed")
        else:
            print(f"❌ Test 1 Failed. Got {result}")
    except Exception as e:
        print(f"⚠️ Test 1 Error: {e}")
    if passed == total:
        print("\n🎉 ALL TESTS PASSED!")
    else:
        print(f"\n📊 Score: {passed}/{total} tests passed")

evaluate()
```
