```python
# --- EVALUATION ENGINE ---
def evaluate():
    passed = 0
    total = 1

    test_cases = [
        ([["1","1","0"],["0","1","0"],["0","0","1"]], 3),
        ([["1","1","1"],["0","1","0"],["1","1","1"]], 1),
    ]
    for i, case in enumerate(test_cases):
        *args, expected = case
        try:
            result = num_islands(*args) if len(args) > 1 else num_islands(args[0])
            if result == expected:
                passed += 1
                print(f"✅ Test {i+1} Passed")
            else:
                print(f"❌ Test {i+1} Failed. Expected {expected}, got {result}")
        except Exception as e:
            print(f"⚠️ Test {i+1} Error: {e}")

    if passed == total:
        print("\n🎉 ALL TESTS PASSED!")
    else:
        print(f"\n📊 Score: {passed}/{total} tests passed")

evaluate()
```
