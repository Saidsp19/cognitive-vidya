```python
# --- EVALUATION ENGINE ---
def evaluate():
    passed = 0
    total = 1

    test_cases = [("A man, a plan, a canal: Panama", True), ("race a car", False), (" ", True)]
    for i, case in enumerate(test_cases):
        *args, expected = case
        try:
            result = is_palindrome(*args) if len(args) > 1 else is_palindrome(args[0])
            if result == expected:
                passed += 1
                print(f"✅ Test {i+1} Passed")
            else:
                print(f"❌ Test {i+1} Failed. Expected {expected}, got {result}")
        except Exception as e:
            print(f"⚠️ Test {i+1} Error: {e}")

    if passed == total:
        print("\n🎉 ALL TESTS PASSED!")
    else:
        print(f"\n📊 Score: {passed}/{total} tests passed")

evaluate()
```
