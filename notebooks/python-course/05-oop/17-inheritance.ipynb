Extend classes with inheritance.

Complete the exercises below, then run the grader cell.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError

class Dog(Animal):
    def speak(self):
        pass

class Cat(Animal):
    def speak(self):
        pass
```

```python
# --- EVALUATION ENGINE ---
def evaluate():
    passed = 0
    total = 1

    if Dog("Rex").speak() == "Rex says Woof!" and Cat("Luna").speak() == "Luna says Meow!":
        passed += 1; print("✅ Inheritance Passed")
    else:
        print("❌ Inheritance Failed")

    if passed == total:
        print("\n🎉 ALL TESTS PASSED! Great work!")
    else:
        print(f"\n📊 Score: {passed}/{total} tests passed")

evaluate()
```
